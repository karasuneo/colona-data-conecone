```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

from sklearn.linear_model import LinearRegression
```

```python
summary_by_date_data = pd.read_csv("../../data/summary_by_date.csv")
# 日付をインデックスとして設定
summary_by_date_data.set_index("date", inplace=True)

# 折れ線グラフを描画
plt.figure(figsize=(12, 6))

# 各列の折れ線グラフを描画
for column in summary_by_date_data.columns:
    plt.plot(summary_by_date_data.index, summary_by_date_data[column], label=column)

plt.title("COVID-19 Vaccination Progress")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
