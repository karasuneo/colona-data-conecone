```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

from sklearn.linear_model import LinearRegression
```

```python
newly_confirmed_cases_daily_data = pd.read_csv(
    "../data/newly_confirmed_cases_daily.csv"
)
severe_cases_daily_data = pd.read_csv("../data/severe_cases_daily.csv")
```

### 欠損値の処理

```python
severe_cases_daily_data.fillna(method="ffill", inplace=True)  # 直前の値で埋める
```

```python
# 日付をインデックスとして設定
newly_confirmed_cases_daily_data["Date"] = pd.to_datetime(
    newly_confirmed_cases_daily_data["Date"]
)
newly_confirmed_cases_daily_data.set_index("Date", inplace=True)

# 折れ線グラフを描画
plt.figure(figsize=(14, 8))  # グラフのサイズを調整
for column in newly_confirmed_cases_daily_data.columns[1:]:  # 'Date'列以外の各列について
    plt.plot(
        newly_confirmed_cases_daily_data.index,
        newly_confirmed_cases_daily_data[column],
        label=column,
    )

plt.title("COVID-19 Cases Over Time")  # グラフのタイトル
plt.xlabel("Date")  # x軸のラベル
plt.ylabel("Number of Cases")  # y軸のラベル
# plt.legend(loc="upper left")  # 凡例の位置を左上に表示
plt.grid(True)  # グリッドを表示
plt.show()  # グラフを表示
```

```python
# 日付をインデックスとして設定
severe_cases_daily_data["Date"] = pd.to_datetime(severe_cases_daily_data["Date"])
severe_cases_daily_data.set_index("Date", inplace=True)

# 折れ線グラフを描画
plt.figure(figsize=(14, 4))  # グラフのサイズを調整
for column in severe_cases_daily_data.columns[1:]:  # 'Date'列以外の各列について
    plt.plot(
        severe_cases_daily_data.index, severe_cases_daily_data[column], label=column
    )

plt.title("COVID-19 Cases Over Time")  # グラフのタイトル
plt.xlabel("Date")  # x軸のラベル
plt.ylabel("Number of Cases")  # y軸のラベル
# plt.legend(loc="upper left")  # 凡例の位置を左上に表示
plt.grid(True)  # グリッドを表示
plt.show()  # グラフを表示
```
